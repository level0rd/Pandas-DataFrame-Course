```python
from pandas import DataFrame
import os
import re

txt = '.txt' 
dir = os.path.abspath(os.curdir)
col_names =  ['Year','M','F','MPR','FPR']
dfres  = pd.DataFrame(columns = col_names)
dfres 

for file in os.listdir(dir):
    if file.endswith(txt):
        digs = re.sub("\D", "", file)
        year = int(digs)
        df_temp = pd.read_csv(file, sep = ",", names=['Name','Sex','Count']) 
        df_temp['Name'] = df_temp['Name'].str.upper()
        dff=df_temp.loc[df_temp.iloc[:,0].str.contains(r'LESL')]
        m = dff.query('Sex == "M"').Sex.count()
        f = dff.query('Sex == "F"').Sex.count()
        total=m+f
        fp=(f/total)*100
        mp=(m/total)*100

        dfres.loc[len(dfres)] = [year,m,f,mp,fp]
```
